### ‚úèÔ∏è M√©todos Auxiliares

````python
def display_dfs(dfs, gap=50, justify='center'):
    html = ""
    for title, df in dfs.items():  
        df_html = df._repr_html_()
        cur_html = f'<div> <h3>{title}</h3> {df_html}</div>'
        html +=  cur_html
    html= f"""
    <div style="display:flex; gap:{gap}px; justify-content:{justify};">
        {html}
    </div>
    """
    display_html(html, raw=True)
````

### ‚ùì Pergunta 1: Quais os meus artistas, √°lbuns e faixas mais ouvidos de todos os tempos?

````python
frequencia = df['nome_do_album'].value_counts()
percentual = df['nome_do_album'].value_counts(normalize = True) * 100
albuns_mais_ouvidos = pd.DataFrame({'Frequ√™ncia': frequencia, 'Porcentagem (%)': percentual})
top_10_albuns = albuns_mais_ouvidos[0:10]

frequencia = df['nome_da_faixa'].value_counts()
percentual = df['nome_da_faixa'].value_counts(normalize = True) * 100
faixas_mais_ouvidas = pd.DataFrame({'Frequ√™ncia': frequencia, 'Porcentagem (%)': percentual})
top_10_faixas = faixas_mais_ouvidas[0:10]

frequencia = df['nome_do_artista'].value_counts()
percentual = df['nome_do_artista'].value_counts(normalize = True) * 100
artistas_mais_ouvidos = pd.DataFrame({'Frequ√™ncia': frequencia, 'Porcentagem (%)': percentual})
top_10_artistas = artistas_mais_ouvidos[0:10]
````

````python
dfs = {'Top 10 Artistas': top_10_artistas, 'Top 10 √Ålbuns': top_10_albuns, 'Top 10 Faixas': top_10_faixas}
display_dfs(dfs, justify='flex-start')
````

````python
top_10_faixas.sort_values(by=['Frequ√™ncia'], inplace=True, ascending=True)
top_10_albuns.sort_values(by=['Frequ√™ncia'], inplace=True, ascending=True)
top_10_artistas.sort_values(by=['Frequ√™ncia'], inplace=True, ascending=True)

fig, (ax1, ax2, ax3) = plt.subplots(3, figsize=(18,14))
fig.suptitle('Estat√≠sticas de Reprodu√ß√£o de Todo o Tempo', fontsize=30)

# Vari√°veis
title_tam = 22
ytitle_tam = 25

ax1.barh(top_10_artistas.index, top_10_artistas['Frequ√™ncia'], color='C13', edgecolor='black')
ax1.set_title('10 Artistas mais Reproduzidos', fontsize =title_tam, pad=5)
ax1.set_ylabel('Artistas', fontsize=ytitle_tam, loc='top', rotation=0)
ax1.yaxis.set_label_coords(-0.01, 0.97)
ax1.tick_params(axis='both', which='major', labelsize=12)
for i, v in enumerate(top_10_artistas['Frequ√™ncia']):
    ax1.text(5, i , str(v) + " Reprodu√ß√µes", color='black', verticalalignment='center')

ax2.barh(top_10_albuns.index, top_10_albuns['Frequ√™ncia'], color='C10', edgecolor='black')
ax2.set_title('10 √Ålbuns mais Reproduzidos', fontsize=title_tam , pad=5)
ax2.set_ylabel('√Ålbuns', fontsize=ytitle_tam, loc='top', rotation=0)
ax2.yaxis.set_label_coords(-0.01, 0.97)
ax2.tick_params(axis='both', which='major', labelsize=12)
for i, v in enumerate(top_10_albuns['Frequ√™ncia']):
    ax2.text(5, i , str(v) + " Reprodu√ß√µes", color='black', verticalalignment='center')

ax3.barh(top_10_faixas.index, top_10_faixas['Frequ√™ncia'], color='C24', edgecolor='black')
ax3.set_title('10 Faixas mais Reproduzidas', fontsize=title_tam , pad=5)
ax3.set_ylabel('Faixas', fontsize=ytitle_tam, loc='top', rotation=0)
ax3.set_xlabel('N√∫mero de Reprodu√ß√µes', fontsize=30)
ax3.yaxis.set_label_coords(-0.01, 0.97)
ax3.tick_params(axis='both', which='major', labelsize=10)
for i, v in enumerate(top_10_faixas['Frequ√™ncia']):
    ax3.text(3, i , str(v) + " Reprodu√ß√µes", color='black', verticalalignment='center')

fig.tight_layout()
#plt.savefig('stats_all_time.png')
````

### ‚ùì Pergunta 5: Quantas m√∫sicas foram ouvidas m√™s a m√™s?

````python
## Algortimo para encontrar a quantidade de m√∫sicas ouvidas por m√™s
ano = 2018
dfs = []
while ano < 2022:
    year_tmp = df[df['ano_de_reprodu√ß√£o'] == ano]
    year_tmp = year_tmp.groupby('m√™s_de_reprodu√ß√£o').nome_da_faixa.count()
    df_temp1 = year_tmp.to_frame()
    df_temp1.columns = ['quantidade']
    df_temp1['m√™s_de_reprodu√ß√£o'] = year_tmp.index
    df_temp1.reset_index(drop=True, inplace=True)
    df_temp2 = df_temp1.sort_values('quantidade', ascending=False)
    df_temp2['ano_de_reprodu√ß√£o'] = ano
    df_temp2.reset_index(drop=True, inplace=True)
    dfs.append(df_temp2)
    ano+=1
    
df_ano_mes = pd.concat(dfs, ignore_index=True)
df_mes_a_mes = df_ano_mes.pivot('m√™s_de_reprodu√ß√£o','ano_de_reprodu√ß√£o','quantidade')
````

````python
df_mes_a_mes
````

````python
## Gerando a visualiza√ß√£o
fig = plt.figure(figsize=(18,6))
ax = fig.add_axes([0, 0, 1, 1])

sns.heatmap(df_mes_a_mes, annot=False, cbar=True, vmin=1, cmap="OrRd")

ax.set_ylabel('M√™s', fontsize=20)
ax.set_xlabel('Ano', fontsize=20)
ax.set_title('Reprodu√ß√£o M√™s a M√™s',fontsize = 30, pad=20)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.show()
````

### ‚ùì Pergunta 6: Qual o artista mais ouvido de cada ano?

Vamos investigar qual o artista mais ouvido de cada ano verificando ano a ano quais os artistas s√£o mais ouvidos, separando o artista mais ouvido encontrado em um DataFrame separado.

````python
## Algoritmo para encontrar o artista mais ouvido de cada ano
ano = 2018
dfs = []
while ano<2022:
    # Seleciona o ano X, come√ßando em 2018 e indo at√© 2021
    temp_ano = df[df['ano_de_reprodu√ß√£o'] == ano]
    # Agrupa a quantidade de artistas desse ano
    temp_ano = temp_ano.groupby('nome_do_artista').nome_do_artista.count()
    # Converte a s√©rie em DataFrame
    df_temp1 = temp_ano.to_frame()
    # Nomeia a coluna quantidade de artistas desse ano como 'quantidade'
    df_temp1.columns=['quantidade']
    # Adiciona a coluna 'nome_do_artista' ao df criado
    df_temp1['nome_do_artista'] = temp_ano.index
    # Reseta o √≠ndice do DataFrame
    df_temp1.reset_index(drop=True, inplace=True)
    # Ordena o DataFrame em ordem ascendente e retorna o artista mais ouvido do ano
    df_temp2 = df_temp1.sort_values('quantidade', ascending=False).head(1)
    # Cria uma coluna com o ano respectivo
    df_temp2['ano_de_reprodu√ß√£o'] = ano
    # Reseta o √≠ndice do DataFrame
    df_temp2.reset_index(drop=True, inplace=True)
    # Adiciona a linha contendo a quantidade de reprodu√ß√µes, o nome do artista e o ano
    # ao DataFrame dfs
    dfs.append(df_temp2)
    # Incrementa o ano
    ano+=1
    
df_artista_mais_ouvido_ano = pd.concat(dfs, ignore_index=True)
df_artista_mais_ouvido_ano.columns = ['quantidade', 'Artista', 'ano_de_reprodu√ß√£o']
````

````python
# Exibe o DataFrame gerado pelo algoritmo
df_artista_mais_ouvido_ano.head()
````

````python
## Gerando a visualiza√ß√£o
g=sns.relplot(
    x="ano_de_reprodu√ß√£o", y="quantidade", hue="Artista", size="Artista",
    sizes =(100, 200), alpha=.9,
    height = 10, 
    aspect = 0.7,
    data = df_artista_mais_ouvido_ano )

g.fig.set_size_inches(10,5)
plt.title("Meus Artistas mais Escutados por Ano", fontsize = 20) 
plt.xlabel('Ano', fontsize=16)
plt.ylabel('N√∫mero de Reprodu√ß√µes', fontsize=16)
plt.show()
````

### ‚ùì Pergunta 7: Qual o √°lbum mais ouvido de cada ano?

Vamos replicar o algoritmo utilizado para determinar o artista mais ouvido de cada ano, mas utilizando as vari√°veis de √°lbum.

````python
## Algoritmo para encontrar o √°lbum mais ouvido de cada ano
ano = 2018
dfs = []
while ano<2022:
    # Seleciona o ano X, come√ßando em 2018 e indo at√© 2021
    temp_ano = df[df['ano_de_reprodu√ß√£o'] == ano]
    # Agrupa a quantidade de √°lbuns desse ano
    temp_ano = temp_ano.groupby('nome_do_album').nome_do_album.count()
    # Converte a s√©rie em DataFrame
    df_temp1 = temp_ano.to_frame()
    # Nomeia a coluna quantidade de artistas desse ano como 'quantidade'
    df_temp1.columns=['quantidade']
    # Adiciona a coluna 'nome_do_album' ao df criado
    df_temp1['nome_do_album'] = temp_ano.index
    # Reseta o √≠ndice do DataFrame
    df_temp1.reset_index(drop=True, inplace=True)
    # Ordena o DataFrame em ordem ascendente e retorna o √°lbum mais ouvido do ano
    df_temp2 = df_temp1.sort_values('quantidade', ascending=False).head(1)
    # Cria uma coluna com o ano respectivo
    df_temp2['ano_de_reprodu√ß√£o'] = ano
    # Reseta o √≠ndice do DataFrame
    df_temp2.reset_index(drop=True, inplace=True)
    # Adiciona a linha contendo a quantidade de reprodu√ß√µes, o nome do √°lbum e o ano
    # ao DataFrame dfs
    dfs.append(df_temp2)
    # Incrementa o ano
    ano+=1
    
df_album_mais_ouvido_ano = pd.concat(dfs, ignore_index=True)
df_album_mais_ouvido_ano.columns = ['quantidade', '√Ålbum', 'ano_de_reprodu√ß√£o']
````

````python
# Exibe o DataFrame gerado pelo algoritmo
df_album_mais_ouvido_ano
````

````python
## Gerando a visualiza√ß√£o
g = sns.relplot(
    x="ano_de_reprodu√ß√£o", y="quantidade", hue="√Ålbum", size="√Ålbum",
    sizes =(100, 200), alpha=.9,
    height = 10, 
    aspect = 0.7,
    data = df_album_mais_ouvido_ano )

g.fig.set_size_inches(10,5)
plt.title("Meus √Ålbuns mais Escutados por Ano", fontsize = 20) 
plt.xlabel('Ano', fontsize=16)
plt.ylabel('N√∫mero de Reprodu√ß√µes', fontsize=16)
plt.show()
````
